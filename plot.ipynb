```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Try Times New Roman, fallback to serif
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 12

# Domain and parameters
L = 1.0
x = np.linspace(0, L, 1000)
mu = L / 2
sigma = 0.01 * L

# Gaussian approximation of Dirac delta
delta_approx = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Plotting
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, delta_approx, color='blue', label=r'Gaussian Approximation of $\delta(x - L/2)$')
ax.axvline(mu, color='black', linestyle='--', linewidth=1, label='Ideal Dirac Delta Location')

# Styling and annotations
ax.set_title('Regularization of Dirac Delta via Gaussian', fontsize=18)
ax.set_xlabel(r'$x$', fontsize=16)
ax.set_ylabel(r'$f(x)$', fontsize=16)
ax.legend(fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Save plot
# plot_path = "/mnt/data/figure_delta_regularization.png"
# plt.savefig(plot_path, dpi=300)
plt.show()
```
